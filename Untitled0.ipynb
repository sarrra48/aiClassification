```python
df1 = pd.read_csv("sample.csv")
df2 = pd.read_csv("accounting_billing_tweets.csv")
df = pd.concat([df1, df2], ignore_index=True)
df.head()
```

```python


print(df.columns)
df.head()
```

```python

df.head()
```

```python
df = df[["text"]]
df.head()
```

```python
def assign_label(text):
    text = str(text).lower()
    if "refund" in text or "payment" in text or "charge" in text or "invoice" in text or "billing" in text or "card" in text:
        return "Billing"
    elif "error" in text or "crash" in text or "bug" in text or "not working" in text or "issue" in text or "problem" in text:
        return "Technical"

    elif "password" in text or "login" in text or "account" in text or "username" in text or "email" in text or "profile" in text:
        return "Account"
    else:
        return "Technical"
```

```python
df["label"] = df["text"].apply(assign_label)
df["label"].value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df["text"], df["label"], test_size=0.2, random_state=42, stratify=df["label"])
```

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
```

```python
model = Pipeline([
    ("TF-IDF", TfidfVectorizer(
        lowercase=True,
        stop_words="english",
        ngram_range=(1,2),
        max_features=5000
    )),
    ("clf", LogisticRegression(
        max_iter=1000,
        class_weight="balanced"
    ))
])
```

```python
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_predict))
```

```python
new_messages = [
    "Why was my credit card charged twice?",
    "The app crashes when I open it",
    "I forgot my account password"
]

predictions = model.predict(new_messages)

for msg, pred in zip(new_messages, predictions):
    print(f"{msg}  {pred}")
```
